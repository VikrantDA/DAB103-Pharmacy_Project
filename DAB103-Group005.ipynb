## Data Cleaning and Manipulation
### Removing Missing values  

```python
merged_df.dropna(inplace = True)
display(merged_df)
```

```python
merged_df = merged_df.drop_duplicates()
display(merged_df)
```

We will be removing **'XX'** Location as it is considered to be _"National Totals"_ and _"Annual State Detail"_ which will not be useful for our analysis as we are focussing on comparision between different regions. Hence, "National Totals" data will be not considered and involved in our analysis report.

```python
#Filter unknown locations
exclude_locations = ['XX']
# Filter the data frame to exclude the specified locations
merged_df = merged_df[~merged_df['state'].isin(exclude_locations)]
```

```python
merged_df['state'].unique()
```

### Categorizing states into major 'Regions' of USA 

```python
def get_region(state):
    if state in ['ME', 'NH', 'VT', 'MA', 'RI', 'CT', 'NY', 'NJ', 'PA']:
        return 'NorthEast'
    elif state in ['OH', 'IN', 'IL', 'MI', 'WI', 'MN', 'IA', 'MO', 'ND', 'SD', 'NE', 'KS']:
        return 'MidWest'
    elif state in ['DE', 'MD', 'DC', 'VA', 'WV', 'NC', 'SC', 'GA', 'FL', 'KY', 'TN', 'AL', 'MS', 'AR', 'LA', 'TX', 'OK']:
        return 'South'
    elif state in ['MT', 'ID', 'WY', 'CO', 'NM', 'AZ', 'UT', 'NV', 'WA', 'OR', 'CA', 'AK', 'HI']:
        return 'West'
    else:
        return 'Other'
merged_df['Region'] = merged_df['state'].apply(lambda x: get_region(x))
display(merged_df)
```

```python
#Filter unknown locations
exclude_locations = [4]
# Filter the data frame to exclude the specified locations
merged_df = merged_df[~merged_df['quarter'].isin(exclude_locations)]
```

Creating seperate dataframes for **MCOU** and **FFSU** - Utilization type by _df_mcou & df_ffsu_ for filtering data.

```python
df_ffsu = merged_df.loc[merged_df['utilization_type'] == 'FFSU']
df_mcou = merged_df.loc[merged_df['utilization_type'] == 'MCOU']
```

```python
display(merged_df)
```

## Plot 1

Plot to represent the distribution of FFS utilization type for all four regions of USA in 2021 and 2022

```python
# filtering the data for the year 2021
df_1 = merged_df[merged_df['year'] == 2021]
ffsu_pie_1 = df_1['Region'].value_counts()

# plot the pie chart for the year 2021 
colors_1 = ['#C0C0C0', '#FFC0CB', '#ADD8E6', '#98AFC7']
plt.pie(ffsu_pie_1, labels=ffsu_pie_1.index, autopct='%1.1f%%', colors=colors_1)
plt.legend(title='Regions', loc='best', bbox_to_anchor=(1, 0, 0.5, 1))
plt.title(f'Number of occurrences of FFSU by Region in 2021')
plt.show()

# filtering the data for the year 2022
df_2 = merged_df[merged_df['year'] == 2022]
ffsu_pie_2 = df_2['Region'].value_counts()

# plot the pie chart for the year 2022 
colors_2 = ['#C0C0C0', '#FFC0CB', '#ADD8E6', '#98AFC7']
plt.pie(ffsu_pie_2, labels=ffsu_pie_2.index, autopct='%1.1f%%', colors=colors_2)
plt.legend(title='Regions', loc='best', bbox_to_anchor=(1, 0, 0.5, 1))
plt.title(f'Number of occurrences of FFSU by Region in 2022')
plt.show()
```

## Plot 2

Plot to represent the distribution of MCO utilization type for all four regions of USA in 2021 and 2022

```python
# filtering the data for the year 2021
df_1 = df_mcou[df_mcou['year'] == 2021]
mcou_pie_1 = df_1['Region'].value_counts()

# plot the pie chart for the year 2021
colors_1 = ['#87AFC7', '#E3E4FA', '#C2B280', '#FFFFCC']
plt.pie(mcou_pie_1, labels=mcou_pie_1.index, autopct='%1.1f%%', colors=colors_1)
plt.legend(title='Regions', loc='best', bbox_to_anchor=(1, 0, 0.5, 1))
plt.title(f'Number of occurrences of MCOU by Region in 2021')
plt.show()

# filtering the data for the year 2022
df_2 = df_mcou[df_mcou['year'] == 2022]
mcou_pie_2 = df_2['Region'].value_counts()

# plot the pie chart for the year 2022
colors_2 = ['#87AFC7', '#E3E4FA', '#C2B280', '#FFFFCC']
plt.pie(mcou_pie_2, labels=mcou_pie_2.index, autopct='%1.1f%%', colors=colors_2)
plt.legend(title='Regions', loc='best', bbox_to_anchor=(1, 0, 0.5, 1))
plt.title(f'Number of occurrences of MCOU by Region in 2022')
plt.show()
```

## Plot 3

Bar Plot to compare the amount reimbursed over all four regions of USA in **2021 and 2022**

```python
df_2021 = merged_df.loc[merged_df['year'] == 2021]
df_2022 = merged_df.loc[merged_df['year'] == 2022]

# count the total amount reimbursed in each region for each year
northeast_count_2021 = df_2021.loc[df_2021['Region'] == 'NorthEast', 'total_amount_reimbursed'].sum()
midwest_count_2021 = df_2021.loc[df_2021['Region'] == 'MidWest', 'total_amount_reimbursed'].sum()
south_count_2021 = df_2021.loc[df_2021['Region'] == 'South', 'total_amount_reimbursed'].sum()
west_count_2021 = df_2021.loc[df_2021['Region'] == 'West', 'total_amount_reimbursed'].sum()

northeast_count_2022 = df_2022.loc[df_2022['Region'] == 'NorthEast', 'total_amount_reimbursed'].sum()
midwest_count_2022 = df_2022.loc[df_2022['Region'] == 'MidWest', 'total_amount_reimbursed'].sum()
south_count_2022 = df_2022.loc[df_2022['Region'] == 'South', 'total_amount_reimbursed'].sum()
west_count_2022 = df_2022.loc[df_2022['Region'] == 'West', 'total_amount_reimbursed'].sum()

# plot the results for 2021 and 2022
region_counts_2021 = pd.Series([northeast_count_2021, midwest_count_2021, south_count_2021, west_count_2021], index=['Northeast', 'Midwest', 'South', 'West'])
region_counts_2022 = pd.Series([northeast_count_2022, midwest_count_2022, south_count_2022, west_count_2022], index=['Northeast', 'Midwest', 'South', 'West'])
colors_2021 = ['#2B3856', '#2B3856', '#2B3856', '#2B3856']
colors_2022 = ['#A2AD9C', '#A2AD9C', '#A2AD9C', '#A2AD9C']

fig, ax = plt.subplots(figsize=(12, 6))
region_counts_2021.plot(kind='bar', color=colors_2021, position=0.5, width=0.4, ax=ax, label='2021')
region_counts_2022.plot(kind='bar', color=colors_2022, position=-0.5, width=0.4, ax=ax, label='2022')
plt.title('Amount reimbursed over Region in 2021 and 2022', fontsize=16, fontweight='bold')
plt.xlabel('Region', fontsize=14)
plt.ylabel('Amount ($)', fontsize=14)
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%d'))
ax.legend()

plt.show()
```

## Plot 5

Bar chart to identify the top five drug prescribed in all four regions of USA

```python
# identifying the count of each drug in each region
merged_df_count = merged_df.groupby(['Region', 'product_name'])['product_name'].count().reset_index(name='Count')

# Sort the data by count in descending order
merged_df_sorted = merged_df_count.sort_values(by='Count', ascending=False)

# Keep only the top 5 drugs for each region
merged_df_top5 = merged_df_sorted.groupby('Region').head(5)

palette = sns.light_palette("#045F5F", 4, reverse=True)

# Plotting the results
g = sns.FacetGrid(merged_df_top5, col='Region', col_wrap=2)
g.map(sns.barplot, 'product_name', 'Count', order=merged_df_top5['product_name'].unique()[:5], palette=palette)
g.set_xticklabels(rotation=90)
plt.show()
```
